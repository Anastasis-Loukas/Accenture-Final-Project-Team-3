```python
dim_region = spark.table("accenture_final_project.silver_layer.region")
display(dim_region)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

window = Window.orderBy("SalesTerritoryKey")

dim_region = dim_region.withColumn(
    "RegionSK",
    row_number().over(window)
)
```

```python
dim_region = dim_region.select("RegionSK","SalesTerritoryKey","Region","Country","Group")
```

```python
display(dim_region)
```

```python
(
    dim_region.write 
         .format("delta") 
         .mode("overwrite") 
         .saveAsTable("accenture_final_project.gold_layer.dim_region")
)
```
