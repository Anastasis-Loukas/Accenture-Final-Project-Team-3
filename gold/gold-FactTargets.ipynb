```python
dim_date = spark.read.format("delta").table("accenture_final_project.gold_layer.dim_date")
dim_salesperson = spark.read.format("delta").table("accenture_final_project.gold_layer.dim_salesperson")
targets = spark.read.format("delta").table("accenture_final_project.silver_layer.targets")
display(targets)
```

```python
#Join the tables to create the fact table
#We use left join to keep all targets 
fact_targets = targets.join(dim_date,targets.TargetDate==dim_date.Date,"left")\
    .join(dim_salesperson,targets.EmployeeID==dim_salesperson.EmployeeID,"left")

#Select the columns we want to include in the fact_targets table
fact_targets = fact_targets.select("EmployeeSK","Target","DateSK")
display(fact_targets)
```

```python
#Save fact_targets table as Delta Table in the gold layer
fact_targets.write.format("delta").mode("overwrite").saveAsTable("accenture_final_project.gold_layer.fact_targets")
```
