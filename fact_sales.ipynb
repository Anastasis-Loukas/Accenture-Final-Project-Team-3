```python
# To make the fact table sales we need the dim_product , dim_date , dim_region , dim_reseller and dim_salesperson tables
# From each of these dimentions we need to get with joins the surrogate keys that they have 

import pyspark.sql.functions as F

dim_date = spark.read.format("delta").table("accenture_final_project.gold_layer.dim_date")
dim_salesperson = spark.read.format("delta").table("accenture_final_project.gold_layer.dim_salesperson")
dim_product = spark.read.format("delta").table("accenture_final_project.gold_layer.dim_product")
dim_region = spark.read.format("delta").table("accenture_final_project.gold_layer.dim_region")
dim_reseller = spark.read.format("delta").table("accenture_final_project.gold_layer.dim_reseller")
sales = spark.read.format("delta").table("accenture_final_project.silver_layer.sales")
```

```python

fact_sales = (
                sales.join(dim_product, sales.ProductKey == dim_product.ProductKey, "left")
                .join(dim_date, sales.OrderDate == dim_date.Date, "left")
                .join(dim_region, sales.SalesTerritoryKey == dim_region.SalesTerritoryKey, "left")
                .join(dim_reseller, sales.ResellerKey == dim_reseller.ResellerKey, "left")
                .join(dim_salesperson, sales.EmployeeKey == dim_salesperson.EmployeeKey, "left")
                .select("DateSK","RegionSK" ,"ProductSK", "ResellerSK", "EmployeeSK","SalesOrderNumber","Quantity", "UnitPrice", "Sales","Cost")
             )

display(fact_sales)


                  
```

```python
fact_sales.write.mode("overwrite").format("delta").saveAsTable("accenture_final_project.gold_layer.fact_sales")
```

```python
%sql
SELECT * FROM accenture_final_project.gold_layer.fact_sales
WHERE ProductSK = null;
```

```python
# Calculating nulls for each column
null_counts = fact_sales.select([
    F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in fact_sales.columns
])

print("Null count per column:")
display(null_counts)
```

```python
# Φιλτράρουμε τον πίνακα για να κρατήσουμε ΜΟΝΟ τις γραμμές όπου το ProductSK είναι null
null_products_df = fact_sales.filter(F.col("ProductSK").isNull())

# Εμφανίζουμε το αποτέλεσμα
print("Γραμμές με null ProductSK:")
display(null_products_df)
```
