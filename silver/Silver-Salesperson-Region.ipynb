## Salesperson Region Table Cleaning

```python
#Display the contents of the salesperson table in the bronze layer
#drop column _resqued_data generated during ingestion in the bronze layer
df_salesperson_region = spark.table("accenture_final_project.bronze_layer.salespersonregion").drop("_rescued_data")
display(df_salesperson_region)
```

```python
from pyspark.sql.types import IntegerType
from pyspark.sql import functions as F

#Define schema just to make sure the schema is correct
df_salesperson_region = df_salesperson_region.select(
    F.col("EmployeeKey").cast(IntegerType()),
    F.col("SalesTerritoryKey").cast(IntegerType())
)
```

```python
#Check for duplicates
duplicates = df_salesperson_region.groupBy(df_salesperson_region.columns).count().filter(F.col("count") > 1)
display(duplicates)
```

```python
#Check for missing values
missing_values = df_salesperson_region.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in df_salesperson_region.columns])
display(missing_values)
```

## Transform to Delta Table

```python
#Save Salesperson_Region DataFrame as Delta Table inside the silver layer
df_salesperson_region.write.mode("overwrite").option("overwriteSchema", "true").format("delta").saveAsTable("accenture_final_project.silver_layer.salespersonregion")
```
