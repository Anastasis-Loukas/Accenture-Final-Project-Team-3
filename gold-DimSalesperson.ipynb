```python
salesperson = spark.read.format("delta").table("accenture_final_project.silver_layer.salesperson")
display(salesperson)
```

```python
from pyspark.sql import Window
from pyspark.sql.functions import row_number

#Create window to generate a surrogate key for each row in the salesperson dimension table
window = Window.orderBy("EmployeeKey")

dim_salesperson = salesperson.withColumn("EmployeeSK",row_number().over(window))

#select columns with specific order to 
dim_salesperson = dim_salesperson.select("EmployeeSK","EmployeeKey", "EmployeeID", "Salesperson", "Title", "UPN")
display(dim_salesperson)
```

```python
#Save as Delta table in the gold layer
dim_salesperson.write.format("delta").mode("overwrite").saveAsTable("accenture_final_project.gold_layer.dim_salesperson")
```
