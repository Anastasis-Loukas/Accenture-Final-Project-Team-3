```python
#Display the contents of the SalespersonRegion.csv file
display(spark.read.csv("/Volumes/workspace/accenture_final_project/lakehouse/bronze/SalespersonRegion.csv", header=True, sep="\t"))
```

```python
#Define the Schema for SalespersonRegion
from pyspark.sql.types import StructType, StructField, IntegerType
salesperson_region_schema = StructType([
    StructField("EmployeeKey", IntegerType(), True),
    StructField("SalesTerritoryKey", IntegerType(), True)
])

#Create DataFrame for SalespersonRegion using predefined schema
df_salesperson_region = spark.read.csv("/Volumes/workspace/accenture_final_project/lakehouse/bronze/SalespersonRegion.csv", header=True, sep="\t", schema=salesperson_region_schema)
```

```python
#Rename Column SalesTerritoryKey to Sales_Territory_Key and EmployeeKey to Employee_Key
from pyspark.sql import functions as F

df_salesperson_region = df_salesperson_region.withColumnRenamed("SalesTerritoryKey", "Sales_Territory_Key")\
    .withColumnRenamed("EmployeeKey", "Employee_Key")

display(df_salesperson_region)
```

```python
#Check for duplicates
duplicates = df_salesperson_region.groupBy(df_salesperson_region.columns).count().filter(F.col("count") > 1)
display(duplicates)
```

```python
#Check for missing values
missing_values = df_salesperson_region.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in df_salesperson_region.columns])
display(missing_values)
```

```python
#Save Salesperson_Region DataFrame as Delta Table
df_salesperson_region.write.mode("overwrite").format("delta").saveAsTable("accenture_final_project.salesperson_region_silver")
```
