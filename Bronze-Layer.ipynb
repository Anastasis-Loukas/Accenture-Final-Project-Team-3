## Region Table

```python
#Define Schema for Region 
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

region_schema = StructType([
    StructField("SalesTerritoryKey", IntegerType(),True),
    StructField("Region", StringType(), True),
    StructField("Country", StringType(), True),
    StructField("Group", StringType(), True)
])

#Create DataFrame for Region
bronze_region_df = spark.read.csv("/Volumes/workspace/default/adventure_works_volume/raw/Region.csv",  header=True, sep="\t", schema=region_schema)

#Save as parquet file
bronze_region_df.write.mode("overwrite").parquet("/Volumes/workspace/default/adventure_works_volume/bronze/region_data")
```
